```python
#Import dependencies
import pandas as pd
import datetime as dt
```

```python

#Read Covid Data
covid_data = pd.read_csv('csvcovid.csv', encoding='latin_1')
#Check data is read in
covid_data.head(1)
```

```python
covid_data = covid_data.drop(["Lat", "Long", "Recovered", "Province/State"], axis=1)
```

```python
covid_data.head(3)
```

```python
covid_data.tail(3)
```

```python
#Assign data to week progression of pandemic
#Set Date column to datetime type
dates = pd.to_datetime(covid_data.Date, yearfirst=True)
#Set first date in data
firstdate = dates[0]
#Parse trough dates to assign weeks
weeks = []
week = 1
i = 0
for date in dates:
    if firstdate <= date <= firstdate + dt.timedelta(days=7):
        weeks.append(week)
        i += 1
    else:
        firstdate = dates[i]
        week += 1
        weeks.append(week)
        i += 1
#Set data to appropriate week
covid_data['Week'] = weeks
#Drop unneeded Date column
covid_data.drop('Date', axis=1, inplace=True)
covid_data.head()
```

```python
#Rename Column
covid_data = covid_data.rename(columns={"Country/Region": "Country"})
covid_data = covid_data.rename(columns={"Confirmed": "Cases"})
```

```python
covid_data.head()
```

```python
#Group data by week and country
covid_group = covid_data.groupby(['Week', 'Country'])
covid_group.count().head()
```

```python
#Aggregate(Total) cases and deaths by country per week
covid_df = pd.DataFrame({'Cases':covid_group.Cases.sum(), 'Deaths':covid_group.Deaths.sum()})
covid_df.head()
```

```python
covid_df.to_csv("cleancovids.csv")
```
